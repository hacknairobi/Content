# Introduction to Pandas

### What is Pandas?

Pandas is one of the most popular Python libraries for Data Science and Analytics. 
It helps us to manage two-dimensional data tables in Python. Eg. Excel or CSV files
Of course, it has many more features. In this pandas tutorial, we will learn the most often used things that you have to know from scratch. 

### Why learn Pandas?
A lot of machine learning and deep learning datasets come in spreadsheets in formats such as excel and CSV. Such data cannot be fed directly into teh algorithm. Instead, it requires to be pre-processed first.

Pre-processing basically involves sorting and filtering only the required columns. Some cleaning may also be performed to remove outliers (these are extreme values) that may skew charts or affect the model during training causing inaccuracies.

# Plotting data
This is generating charts inside jupyter using Matplotlib


```python
# Import MatplotLib
import matplotlib.pyplot as plt

# Let matplotlib plot the charts in teh cell
%matplotlib inline

# Select a year
latest = olympics[(olympics.Edition == 2008)]
```

### Line graphs
These are better for representing continous data

```python
# Number of women participating in olympics in Kenya over the years
women = olympics[olympics.NOC.str.contains('KEN') & (olympics.Gender == 'Women')]
women.groupby('Edition').size().plot();
```

### Bar plot

```python
# Plot the number of medals won by men in athletics in Kenya over the years
women = olympics[olympics.NOC.str.contains('KEN') & (olympics.Gender == 'Men') & (olympics.Sport == 'Athletics')]
women.groupby('Athlete').size().plot(kind='bar',figsize=(20,5));
```

# Horizontal Chart

```python
women.groupby('Athlete').size().plot(kind='barh',figsize=(5,20));
```

### Aggregate
Allows multiple statistics to be calculated per group at once

```python
# Group by year, then by country then by the medal and sort them alphabetically
olympics.groupby(['Edition', 'NOC', 'Medal']).agg(['min', 'max', 'count'])
```

```python
# Cleaner, more sensible ouput of the prevoius query
olympics.groupby(['Edition', 'NOC', 'Medal']).agg('count')
```

```python
# Cleaner, more sensible ouput of the prevoius query
olympics.groupby(['Edition', 'NOC', 'Medal']).agg(['min', 'max','count'])
```

```python

# Cleaner, more sensible ouput of the prevoius query
olympics.groupby(['Edition', 'NOC', 'Medal']).size()
```

```python
# Cleaner, more sensible ouput of the prevoius query
olympics.groupby(['Edition', 'NOC', 'Medal']).agg({'Edition':['count']})
```

```python
# Find out when he first won a medal and the number of methods won
olympics.loc[olympics.Athlete == 'NDEREBA, Catherine'].groupby('Athlete').agg({'Edition':['min', 'max', 'count']})
```

### Stack and Unstack
Move inner columns to the rows and vice-versa

```python
# Atheltes that won medals in Beijing Olympics 100m/200m track
MenAndWomen = olympics[(olympics.Edition == 2008) & ((olympics.Event == '100m') | (olympics.Event == '200m'))]
MenAndWomen
```

```python
# Group them by the country, gender, disciplne and the event
grouped = MenAndWomen.groupby(['NOC', 'Gender', 'Discipline', 'Event']).size()
grouped
```

### Stacking
Returns a dataframe or a series
Automatically drops null values

### Unstacking
Returns a dataframe or a series
Null values can prefilled

```python
# Unstack to get the view of the original dataframe
df = grouped.unstack(['Gender'])
df
```

```python
df.stack()
```

### Plot the 5 athletes who have won the most gold medals over the history of the Olympics. When there is a tie, consider the number of silver medals, then bronze medals

```python
# Find
g = olympics.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']].head().plot(kind='barh', figsize=(7,5));
```

### Plot the number of gold medals won by the Kenya male and female athletes throughout the history of the Olympics?

```python
gu =olympics[(olympics.NOC == 'KEN') & (olympics.Medal == 'Gold')]
gu.groupby(['Edition','Gender']).size().unstack('Gender',fill_value=0).plot();
```
