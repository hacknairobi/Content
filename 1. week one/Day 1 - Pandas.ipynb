# Introduction to Pandas

### What is Pandas?

Pandas is one of the most popular Python libraries for Data Science and Analytics. 
It helps us to manage two-dimensional data tables in Python. Eg. Excel or CSV files
Of course, it has many more features. In this pandas tutorial, we will learn the most often used things that you have to know from scratch. 

### Why learn Pandas?
A lot of machine learning and deep learning datasets come in spreadsheets in formats such as excel and CSV. Such data cannot be fed directly into teh algorithm. Instead, it requires to be pre-processed first.

Pre-processing basically involves sorting and filtering only the required columns. Some cleaning may also be performed to remove outliers (these are extreme values) that may skew charts or affect the model during training causing inaccuracies.

### Value counts
It returns a series object counting all the unique values in an object.

Defaults: dropna=true ascending=False

```python
olympics['Edition'].value_counts(ascending=True)
```

### Sort Values
Sorts data in a series aling the column and in ascending order by default.
Missing data is put at the end. 
Defaults: inplace=False, kind='quicksort', na_position='last' axis='0'

```python
# Sort values by year
olympics.Edition.sort_values()
```

```python
# Create a new dataframe and print it
newdataframe = olympics.sort_values(by=['Edition', 'Athlete'])
newdataframe
```

### Boolean indexing

Pass data where the series has true values from a collection of true and false values

```python
olympics.Medal == "Gold"
```

```python
# to see the athletes with a Gold medal

olympics[olympics.Medal == "Gold"]
```

```python
# To input multiple conditions

olympics[(olympics.Medal == 'Gold') & (olympics.Edition == 2008)]
```

### String handling

```python
# Find all players from Kenya
olympics[olympics.NOC.str.contains('KEN')]
```

```python
# Find all athletes that start with 'j'
olympics[olympics.Athlete.str.startswith('J')]
```

# Plotting data
This is generating charts inside jupyter using Matplotlib


```python
# Import MatplotLib
import matplotlib.pyplot as plt

# Let matplotlib plot the charts in teh cell
%matplotlib inline

# Select a year
latest = olympics[(olympics.Edition == 2008)]
```

### Line graphs
These are better for representing continous data

```python
# Plot the number of women who participated in athletics in Kenya over the years
women = latest[latest.NOC.str.contains('KEN') & (latest.Sport == 'Athletics') & (latest.Gender == 'Men')]
# women = women.Athlete.value_counts()
women.Edition.plot(kind='line');
```

### Bar plot

```python
# Plot a line graph showing the number of medals awarded to Kenya over the years
kenyans = olympics[olympics.NOC.str.contains('KEN')]
kenyans.Edition.plot(kind='line');
# kenyans
```

# Horizontal Chart

```python
# Chart showing all the sports Kenyan athletes participated in and their numbers

kenyansports = olympics[olympics.NOC.str.contains('KEN')]
kenyansports.Sport.value_counts().plot(kind='barh');
```

### Pie chart

```python
# Show the distribution of sports to different athletes worldwide in 2008

latest.Sport.value_counts().plot(kind='pie');
```

### Plot colors

```python
# Add a maroon color to our barh chart

kenyansports = olympics[olympics.NOC.str.contains('KEN')]
kenyansports.Sport.value_counts().plot(kind='barh', color='maroon');
```

## Figure sizes

```python
# Change the size of our pie chart(radius)
latest.Sport.value_counts().plot(kind='pie', radius=(3), figsize=(2.5, 2.5));
```

```python
# Space our bar graph of number of participants against each sport
latest.Sport.value_counts(ascending = True).plot(kind='barh', figsize=(10,10)); 
```

### Groupby()
Used to group dataframes by a certainn type before combining them to one

```python
# Group By Country
countries = olympics.groupby('NOC')
list(countries)
```

```python
### Iterating through groups
```
